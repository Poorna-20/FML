```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def tanh(x):
  return np.tanh(x)
```

```python
x = np.linspace(-10 , 10)
print(x)
```

```python
plt.plot(x,tanh(x))
plt.axis('tight')
plt.title("Activation function : Sigmoid")
plt.show()
```
